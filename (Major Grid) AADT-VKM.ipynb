### Section 3: Missing or Duplication Data

```python
#Check for missing value
df1.isnull().sum()
```

```python
#check for data mean and count 
df1.describe()
```

##### Filling in missing values

```python
# Drop rows= 87996 to 87998
df1_label = df1.iloc[87996:87999,:]
df2=df1.drop(list(df1_label.index.values))
df2
```

```python
df2['Year']=2013                                                                 #Fill Year empty content with "2013"        
df2['Toid'].fillna((df1['Toid'].mean()),inplace=True)                            #Fill Toid with mean values
df2['Location_ExactCut'].fillna(method="ffill",inplace=True)                     #Fill Location_ExactCut with ffill method  
df2['BoroughName_ExactCut'].fillna(method="ffill",inplace=True)                  #Fill BoroughName_ExactCut with ffill method
df2['GRID_ExactCut_ID'].fillna((df1['GRID_ExactCut_ID'].mode()[0]),inplace=True) #Fill GRID_EXACTCUT_ID with mode values 
df2['TLRN'].fillna((df1['TLRN'].mode()[0]),inplace=True)                         #Fill TLRN with mode values 
df2['MotorwayNumber'].fillna((df1['MotorwayNumber'].mode()[0]),inplace=True)     #Fill MotorwayNumber with mode values 
```

```python
df2
```

```python
df2.isnull().sum()
```

```python
df2.describe()
```

```python
# Replacing missing values with mean
dfn2= df2.iloc[:,8:]
dfn3= dfn2.mean(axis=0, skipna=True)
dfn4=df2.replace(0,dfn3)
df3=dfn4
```

##### Correlation Performance check

```python
#Check correlation performance
data=pd.DataFrame(df3.iloc[:,8:])
data1=data.corr()
seaborn.heatmap(data1)
```

```python
# Drop those parameters with correlation of greater than 0.90
df3.drop("VKM_Rigid2Axle", axis=1, inplace=True)
df3.drop("VKM_ElectricLgv", axis=1, inplace=True)
df3.drop("AADT_Dcar", axis=1, inplace=True)
df3.drop("VKM_Artic3Axle", axis=1, inplace=True)
df3.drop("AADT_DLgv", axis=1, inplace=True)
```

##### Split data by zone

```python
# Zone: Outer
df4 = df3.loc[df3['Location_ExactCut'] == "Outer"]
# Zone: External
df5= df3.loc[df3['Location_ExactCut']== "External"]
# Zone: Inner
df6= df3.loc[df3['Location_ExactCut']== "Inner"]
# Zone: Central 
df7= df3.loc[df3['Location_ExactCut']== "Central"]
```

##### Section 4: Spliting the data by zone

```python
PATH = "c:\Documents"
OUTPUT_ZONE_FILE="AADT-VKM.xlsx"

FILE_EXIST = os.path.exists(os.path.join(PATH, OUTPUT_ZONE_FILE))
if FILE_EXIST:
    os.remove(os.path.join(PATH, OUTPUT_ZONE_FILE))

with pd.ExcelWriter(os.path.join(PATH, OUTPUT_ZONE_FILE)) as writer:
    df3.to_excel(writer, sheet_name='AADT')
    df4.to_excel(writer, sheet_name='AADT_Central')
    df5.to_excel(writer, sheet_name='AADT_External')
    df6.to_excel(writer, sheet_name='AADT_Inner')
    df7.to_excel(writer, sheet_name='AADT_Outer')
```
